```python
cap = cv2.VideoCapture('./inputVideo/jump_input.mp4')
success,prevFrame = cap.read()
sizeOfFrame = prevFrame.size
print(prevFrame.shape)
plt.imshow(prevFrame)
count = 0
while(success and count<25 ):
    success, curFrame = cap.read()
    if(not success):
        break
    print(curFrame.shape)
    print(curFrame[:,:,0].shape)
    print(curFrame[:,:,0])
#     print(signal.correlate2d(curFrame[:,:,0], prevFrame[:,:,0], mode="valid"))
    corr_b = signal.correlate2d(curFrame[:,:,0], prevFrame[:,:,0],mode="valid")
    corr_g = signal.correlate2d(curFrame[:,:,1], prevFrame[:,:,1], mode="valid")
    corr_r = signal.correlate2d(curFrame[:,:,2], prevFrame[:,:,2], mode="valid")
    count+=1
    print(corr_b, corr_g, corr_r)
    prevFrame = curFrame
```
