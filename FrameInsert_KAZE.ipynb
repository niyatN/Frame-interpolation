```python
import cv2
import os
import matplotlib.pyplot as plt

import scipy
from scipy import signal
import numpy as np
from IPython.display import clear_output
%matplotlib inline
```

```python
def extract_features(image, vector_size=64):
    try:
        alg = cv2.KAZE_create()
        kps = alg.detect(image)
        kps = sorted(kps, key=lambda x: -x.response)[:vector_size]
        kps, dsc = alg.compute(image, kps)
        no_of_kps = len(kps)
        needed_size = (vector_size * 64)
        size = no_of_kps*64
        if(no_of_kps!=0):
#             print(dsc.size)
            dsc = dsc.flatten()
            if(size < needed_size):
                dsc = np.concatenate([dsc, np.zeros(needed_size - size)])
        else:
            dsc = np.zeros(needed_size - size)
#         print(dsc)
#         print(dsc.size,'----------')
    except cv2.error as e:
        print('Error: ', e)
        return None
    return np.reshape(dsc, (1,-1))
```

```python
cap = cv2.VideoCapture('./inputVideo/jump_input.mp4')
success,prevFrame = cap.read()
sizeOfFrame = prevFrame.size

height, width, layers = prevFrame.shape
s = (width, height)
prev_dsc = extract_features(prevFrame)
out = cv2.VideoWriter('./data/outputVideo/6.avi', cv2.VideoWriter_fourcc(*'DIVX'),48,s)
out.write(np.uint8(prevFrame))
while(success):
    success, curFrame = cap.read()
    if(not success):
        break
    cur_dsc = extract_features(curFrame)
    img_dist = scipy.spatial.distance.cdist(prev_dsc,cur_dsc, 'euclidean').reshape(-1)
    prevFrame = curFrame
    prev_dsc = cur_dsc
    out.write(np.uint8(prevFrame))
    if(img_dist<=1):
        out.write(np.uint8(prevFrame/2+curFrame/2))
    print(img_dist)
out.release()
```
