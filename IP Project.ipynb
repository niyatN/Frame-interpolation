```python
import cv2 
import os
```

```python
# Function to extract frames 
def FrameCapture(path): 
    if(not os.path.exists(path)):
        return "FILE_NOT_EXIST"
    
    vidObj = cv2.VideoCapture(path)
    count = 0
    success = 1
  
    while success: 
        success, image = vidObj.read() 
        cv2.imwrite("./data/inputFrame/%d.jpg" % count, image) 
  
        count += 2
```
